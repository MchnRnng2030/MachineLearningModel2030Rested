```python
# ======================================================
# STEP 1: 정책 매트릭스 구조 점검
# ======================================================

import pandas as pd

POLICY_PATH = "../results/oxju/policy_matrix_final.csv"

print("=" * 70)
print("🔍 STEP 1: 정책 매트릭스 점검")
print("=" * 70)

df = pd.read_csv(POLICY_PATH)

print("📊 shape:", df.shape)
print("\n📌 columns:")
for c in df.columns:
    print(" -", c)

# 핵심 컬럼 체크
REQUIRED_COLS = [
    "정책유형",
    "단절위험_환경평균",
    "취업예측확률",
    "재연결지수",
    "지원필요도점수"
]

missing = [c for c in REQUIRED_COLS if c not in df.columns]
if missing:
    raise ValueError(f"❌ 정책 핵심 컬럼 누락: {missing}")

print("\n✅ 정책 매트릭스 정상")
```
