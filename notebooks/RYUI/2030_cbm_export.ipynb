```python
# ======================================================
# CBM 결과 추출 (X_test.csv 사용, FINAL)
# ======================================================

import pandas as pd
import numpy as np
from catboost import CatBoostClassifier

# ------------------------------------------------------
# 1️⃣ 테스트 데이터 로드
# ------------------------------------------------------

X_test = pd.read_csv("X_test.csv")

# pid 없으면 생성 (⚠️ 정책 매트릭스와 규칙 통일!)
if "pid" not in X_test.columns:
    X_test = X_test.reset_index(drop=True)
    X_test["pid"] = X_test.index + 1

# 모델 입력용 feature
X_test_feat = X_test.drop(columns=["pid"])

# ------------------------------------------------------
# 2️⃣ 학습된 CatBoost 모델 로드
# ------------------------------------------------------

cat = CatBoostClassifier()
cat.load_model("cat_model.cbm")

# ------------------------------------------------------
# 3️⃣ 예측
# ------------------------------------------------------

cbm_prob = cat.predict_proba(X_test_feat)[:, 1]

# threshold (이미 정해진 값 있으면 그걸로)
THRESHOLD = 0.5
cbm_pred = (cbm_prob >= THRESHOLD).astype(int)

# ------------------------------------------------------
# 4️⃣ 결과 저장
# ------------------------------------------------------

df_cbm = pd.DataFrame({
    "pid": X_test["pid"],
    "cbm_p_success": cbm_prob,
    "cbm_pred": cbm_pred
})

df_cbm.to_csv("cbm_predictions.csv", index=False)
print("✅ cbm_predictions.csv 생성 완료")
```
