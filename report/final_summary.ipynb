```python
import json
import pandas as pd
from glob import glob

# ======================================================
# 집계용 결과 JSON만 로드 (_result.json)
# ======================================================

files = glob("../results/**/*_result.json", recursive=True)
print("로드된 파일:", files)

rows = []

for f in files:
    with open(f, encoding="utf-8") as fp:
        data = json.load(fp)

    rows.append({
        "owner": data.get("owner", "-"),
        "experiment_id": data.get("experiment_id", "-"),
        "persona": data.get("persona", "-"),
        "hypothesis": data.get("hypothesis", "-"),
        "accuracy": data.get("accuracy", "-"),
        "roc_auc": data.get("roc_auc", "-"),
        "policy_message": data.get("policy_message", "-")
    })

df_results = pd.DataFrame(rows)

# 컬럼 순서 고정
df_results = df_results[
    ["owner", "experiment_id", "persona", "hypothesis",
     "accuracy", "roc_auc", "policy_message"]
]

df_results


# 정책 설계 뷰
```
